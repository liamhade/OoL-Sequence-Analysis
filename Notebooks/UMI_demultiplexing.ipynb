Below are the sequences that were duplicated because of PCR, and which were removed from our DataFrame.

```python
duplicates_df = df_4N_VCG[df_4N_VCG.duplicated(subset=["VCG", "MADAP_Code", "UMI"])]
duplicates_df
```

The vast majority of the duplicated sequences were only copied once. One seq was accidentally copied 37 times during PCR. Further analysis and data is needed to confirm whether or not this sequence is favored by PCR amplification in general.

```python
duplicated_count = duplicates_df[["VCG", "MADAP_Code",  "UMI"]].value_counts()
tags = ["_".join(t) for t in duplicated_count.axes[0]]
lens = [len(t) for t in tags]
As = [t.count("A") for t in tags]
Cs = [t.count("C") for t in tags]
Gs = [t.count("G") for t in tags]
Ts = [t.count("T") for t in tags]

zipped = pd.DataFrame(data={"tag": tags, "lengths": lens, "duplication_freq": duplicated_count, "As": As, "Cs": Cs, "Gs": Gs, "Ts": Ts,})
```
